```python
#desired schema for contributions: 
# PK id
# candidate_id (FK for candidates table)
# contributor_id (FK for contributors table)
# date (taken from contribution_receipt_date)
# amount (taken from contribution_receipt_amount)

# Create contributions table with auto-incrementing ID
# Reset index of contributions first to avoid duplicate index issues
contributions = contributions.reset_index(drop=True)

contributions_f = pd.DataFrame()
contributions_f['id'] = range(len(contributions))
contributions_f['candidate_id'] = contributions['committee_id']
contributions_f['contributor_id'] = contributions['contributor_name']
contributions_f['date'] = pd.to_datetime(contributions['contribution_receipt_date'])
contributions_f['amount'] = contributions['contribution_receipt_amount']
```

```python
# desired schema for contributors table: 
# contributor_id (PK)
# state_id
# first_name
# last_name
# zipcode
# employer
# occupation


contributors= pd.DataFrame()


contributors['contributor_id'] = contributions['contributor_name'] 
contributors['state_id'] = contributions['contributor_state']
contributors['first_name'] = contributions['contributor_first_name']
contributors['last_name'] = contributions['contributor_last_name']
contributors['zipcode'] = contributions['contributor_zip']
contributors['employer'] = contributions['contributor_employer'] 
contributors['occupation'] = contributions['contributor_occupation']


contributors = contributors.drop_duplicates(subset=['contributor_id'])


contributors = contributors.reset_index(drop=True)
```
